````python
# exponentiation modulaire
def modular_pow(base, exponent, modulus):
    result = 1
    base = base % modulus
    while exponent > 0:
        if (exponent % 2 == 1):
            result = (result * base) % modulus
        exponent = exponent >> 1
        base = (base * base) % modulus
    return result

def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

# inverse multiplicatif de a modulo m
def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception("Pas d'inverse multiplicatif")
    else:
      return x % m
````

````python
# Convertir un entier en une chaîne de caractères
def int_to_str(n):
    # Convertir l'entier en une chaîne binaire
    bin_str = bin(n)[2:]  # On enlève le préfixe '0b'

    # S'assurer que la longueur de la chaîne binaire est un multiple de 8
    if len(bin_str) % 8 != 0:
        bin_str = '0' * (8 - len(bin_str) % 8) + bin_str

    # Diviser la chaîne binaire en groupes de 8 bits
    byte_list = [bin_str[i:i + 8] for i in range(0, len(bin_str), 8)]

    # Convertir chaque groupe de 8 bits en un entier, puis en un caractère
    result = ''.join([chr(int(byte, 2)) for byte in byte_list])

    return result
````

###  2- Utilisation de la base de donnees https://fr.wikipedia.org/wiki/ et extraction directe de l'auteur

````python
import requests
from bs4 import BeautifulSoup
import time

def get_wikipedia_authors():
    # URL de base pour chaque page alphabétique des écrivains
    base_url = "https://fr.wikipedia.org/wiki/Liste_alphab%C3%A9tique_d%27%C3%A9crivains_de_langue_fran%C3%A7aise#{}"
    # Liste des lettres pour naviguer sur chaque page
    letters = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
    
    authors = []

    for letter in letters:
        print(f"Extraction des auteurs pour la lettre : {letter}")
        url = base_url.format(letter)
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Erreur lors de la récupération de la page pour la lettre {letter}")
            continue

        # Analyser la page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extraire les noms d'écrivains de la page, en utilisant la balise appropriée
        author_list = soup.select("div.mw-parser-output ul li a")  # Les auteurs sont souvent listés dans des éléments <li>
        for author in author_list:
            name = author.get_text()
            authors.append(name)
        
        # Pause pour éviter de surcharger le serveur
        time.sleep(1)

    return authors

# Utiliser la fonction pour extraire les noms
authors = get_wikipedia_authors()
print(f"Nombre total d'auteurs extraits : {len(authors)}")
print(authors[:10])  # Affiche les 10 premiers auteurs pour vérification
````

### Méthode combinant les Chaines de Markov à l'analyse fréquentielle:

````python
import random as rnd
from collections import Counter, defaultdict
import requests

# Charger et analyser le corpus
def load_text_from_web(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text.upper()  # Mettre tout le texte en majuscules pour uniformité
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors du chargement du texte : {e}")
        return None

# Fonction pour découper le texte en paires de caractères
def cut_string_into_pairs(text):
    pairs = []
    for i in range(0, len(text) - 1, 2):
        pairs.append(text[i:i + 2])
    if len(text) % 2 != 0:
        pairs.append(text[-1] + '_')  # Ajout d'un caractère de remplissage si impair
    return pairs

# Créer un modèle de chaîne de Markov basé sur les bigrammes du corpus
def build_markov_chain(text):
    markov_chain = defaultdict(Counter)
    for i in range(len(text) - 1):
        current_char = text[i]
        next_char = text[i + 1]
        markov_chain[current_char][next_char] += 1
    # Convertir en probabilités
    for current_char, transitions in markov_chain.items():
        total = sum(transitions.values())
        for next_char in transitions:
            transitions[next_char] /= total
    return markov_chain

# Fonction pour générer la clé de substitution
def gen_key(symboles):
    l = len(symboles)
    if l > 256:  
        return False
    rnd.seed(1337)
    int_keys = rnd.sample(list(range(l)), l)
    dictionary = dict({s: "{:08b}".format(k) for s, k in zip(symboles, int_keys)})
    return dictionary

# Fonction pour diviser le message en symboles
def M_vers_symboles(M, K):
    encoded_text = []
    i = 0
    while i < len(M):
        if i + 1 < len(M) and (pair := M[i] + M[i + 1]) in K:
            encoded_text.append(pair)
            i += 2
        elif M[i] in K:
            encoded_text.append(M[i])
            i += 1
        else:
            encoded_text.append('_')  # Placeholder si non trouvé
            i += 1
    return encoded_text

# Fonction pour chiffrer le texte
def chiffrer(M, K):
    l = M_vers_symboles(M, K)
    return ''.join([K.get(x, '_') for x in l])

# Calculer les fréquences des blocs de bits dans le texte chiffré
def calculate_cipher_frequencies(C1):
    cipher_blocks = [C1[i:i + 8] for i in range(0, len(C1), 8)]
    cipher_freq = Counter(cipher_blocks)
    total_blocks = sum(cipher_freq.values())
    return {block: freq / total_blocks for block, freq in cipher_freq.items()}

# Déchiffrer en utilisant une clé hypothétique
def decrypt_with_hypothetical_key(C1, hypothetical_key):
    decrypted_text = []
    for i in range(0, len(C1), 8):
        binary_sequence = C1[i:i+8]
        symbol = hypothetical_key.get(binary_sequence, '?')
        decrypted_text.append(symbol)
    return ''.join(decrypted_text)

# Comparer les fréquences et les transitions pour établir une correspondance
def generate_hypothetical_key(cipher_freq, sorted_symbol_freq, markov_chain):
    hypothetical_key = {}
    used_symbols = set()
    for cipher_block in cipher_freq.keys():
        best_match = None
        best_score = -float('inf')
        for symbol, frequency in sorted_symbol_freq:
            if symbol in used_symbols:
                continue
            # Calcule un "score" pour chaque symbole basé sur la fréquence et la transition
            symbol_freq_score = frequency
            transition_score = sum(markov_chain[symbol].values())  # Plus de poids pour les symboles à forte transition
            score = symbol_freq_score + transition_score
            if score > best_score:
                best_score = score
                best_match = symbol
        hypothetical_key[cipher_block] = best_match or '?'
        if best_match:
            used_symbols.add(best_match)
    return hypothetical_key

# Charger le texte du corpus et créer la chaîne de Markov et les fréquences
urls = [
    "https://www.gutenberg.org/ebooks/13846.txt.utf-8",  # Discours de la méthode - Descartes
    "https://www.gutenberg.org/ebooks/4650.txt.utf-8",   # Candide - Voltaire
]
corpus = ''.join(load_text_from_web(url) for url in urls if load_text_from_web(url))
markov_chain = build_markov_chain(corpus)

# Générer la clé de substitution
K = gen_key(list(set(corpus)))
M1 = corpus[10000:10100]
C1 = chiffrer(M1, K)
cipher_freq = calculate_cipher_frequencies(C1)
symbol_freq = Counter(M_vers_symboles(corpus, set(corpus)))
sorted_symbol_freq = sorted(symbol_freq.items(), key=lambda x: x[1], reverse=True)

# Générer et tester la clé hypothétique
hypothetical_key = generate_hypothetical_key(cipher_freq, sorted_symbol_freq, markov_chain)
decrypted_text = decrypt_with_hypothetical_key(C1, hypothetical_key)

# Debugging outputs
print(f"Corpus length: {len(corpus)}")
print("Sample of Markov chain transitions:")
for k, v in list(markov_chain.items())[:5]:  # Print first 5 for brevity
    print(f"{k}: {dict(v)}")
print("\nSymbol frequencies in corpus (top 5):", sorted_symbol_freq[:5])
print("\nCipher frequencies in encrypted text (top 5):", list(cipher_freq.items())[:5])
print("\nHypothetical key mappings (sample 5):", list(hypothetical_key.items())[:5])

print(f"\n original : {M1}")

print(f"\nTexte déchiffré hypothétique : {decrypted_text}")
````

### Méthode par critére Metropolis et substitution recuit_simule:

````python
import random as rnd
import math
from collections import Counter

# Fonction pour évaluer la vraisemblance du texte déchiffré
def logscore(text_dechiffre):
    bigrams = Counter([text_dechiffre[i:i + 2] for i in range(len(text_dechiffre) - 1)])
    score = sum(bigrams.values())
    return -score

# Critère de Metropolis pour accepter une transition
def critere_Metropolis(delta, T):
    if delta > 0:
        return True
    else:
        return rnd.random() < math.exp(delta / T) if T > 1e-10 else False  # Avoid division by zero

# Génère un nouvel état de substitution en permutant deux éléments
def nouvel_etat(alpha):
    keys = list(alpha.keys())
    i, j = rnd.sample(keys, 2)
    alpha[i], alpha[j] = alpha[j], alpha[i]
    return alpha

# Fonction pour déchiffrer le texte en utilisant la clé alpha
def dechiffrer(text_chiffre, alpha):
    inverse_alpha = {v: k for k, v in alpha.items()}  # Inverser la clé pour le déchiffrement
    return ''.join(inverse_alpha.get(block, '') for block in text_chiffre)

# Recuit simulé pour optimiser la clé de substitution
def recuit_simule_substitution(text_chiffre, symboles, max_iter=8000, cool_ratio=0.8):
    # Génération de la clé initiale aléatoire en permutant les symboles
    rnd.seed(1337)
    shuffled_symbols = rnd.sample(symboles, len(symboles))
    alpha = {sym: shuffled_symbols[i] for i, sym in enumerate(symboles)}
    
    # Déchiffrer le texte chiffré initial pour calculer le score de départ
    text_dechiffre = dechiffrer(text_chiffre, alpha)
    old_score = logscore(text_dechiffre)

    # Calibration de la température initiale
    T = 1e-5  # Initialize to a small value to prevent division by zero
    for _ in range(100): 
        nalpha = nouvel_etat(alpha)
        text_dechiffre = dechiffrer(text_chiffre, nalpha)
        new_score = logscore(text_dechiffre)
        delta = new_score - old_score
        if delta > T: T = delta
        old_score = new_score
    T = max(T * 10, 1e-5)  # Ensure T is large enough for meaningful annealing

    best_alpha = alpha
    best_score = old_score
    freeze = 0

    while True:
        nb_iter = 0
        while nb_iter < max_iter:
            nalpha = nouvel_etat(alpha)
            text_dechiffre = dechiffrer(text_chiffre, nalpha)
            new_score = logscore(text_dechiffre)
            delta = new_score - old_score

            if critere_Metropolis(delta, T): 
                alpha = nalpha
                old_score = new_score
                if old_score < best_score:
                    best_score = old_score
                    best_alpha = alpha
                    print(f"Best score: {best_score}")
                    print(text_dechiffre[:100])  # Affiche un extrait du texte déchiffré
                    print('------')
                freeze = 0
            else:                            
                freeze += 1
            nb_iter += 1

        T *= cool_ratio
        if T < 1e-5:  # Prevent T from approaching zero
            T = 1e-5
        if freeze > 25:
            break

    print("*** Meilleure solution ***")
    print(f'Score = {best_score}')
    print('Texte déchiffré:')
    print(dechiffrer(text_chiffre, best_alpha))

# Generate the key
K = gen_key(symboles)

# Take a segment from the corpus
M1 = corpus[10000:10100]  # Exemple de message

# Encrypt the message segment
C1 = chiffrer(M1, K)  # Corrected to use M1 instead of M
print("Texte chiffré (extrait) :")
print(C1)  # Affiche un extrait

# Lancer le recuit simulé pour déchiffrer
recuit_simule_substitution(C1, symboles)
````

## Clé de chiffrement connue (table de substitution fournie)

````python
import requests
from collections import Counter
import random as rnd

# Generate pairs of characters for bicaracteres
def cut_string_into_pairs(text):
    pairs = []
    for i in range(0, len(text) - 1, 2):
        pairs.append(text[i:i + 2])
    if len(text) % 2 != 0:
        pairs.append(text[-1] + '_')  # Placeholder if odd characters
    return pairs

# Generer une clé de decryptage en inversant le dictionnaire de la clé de chiffrement 

# Fonction pour inverser la clé pour le déchiffrement

def gen_decryption_key(encryption_key):
    return {v: k for k, v in encryption_key.items()}

# Fonction de décryptage pour transformer un ciphertexte binaire en texte clair (plaintext)
def decrypt(C, decryption_key):
    decrypted_text = []
    for i in range(0, len(C), 8):  # Traiter 8 bits d'un coup
        binary_sequence = C[i:i+8]
        symbol = decryption_key.get(binary_sequence, None)
        if symbol:
            decrypted_text.append(symbol)
    return ''.join(decrypted_text)

# Sample data for text processing
url = "https://www.gutenberg.org/ebooks/13846.txt.utf-8"
corpus = requests.get(url).text
caracteres = list(set(list(corpus)))
nb_caracteres = len(caracteres)
nb_bicaracteres = 256 - nb_caracteres
bicaracteres = [item for item, _ in Counter(cut_string_into_pairs(corpus)).most_common(nb_bicaracteres)]
symboles = caracteres + bicaracteres

# Function to generate a random encryption key for the symbols
def gen_key(symboles):
    rnd.seed(1337)
    int_keys = rnd.sample(range(256), len(symboles))
    return {s: "{:08b}".format(k) for s, k in zip(symboles, int_keys)}

# Simulate encryption and decryption for a given message M



def simulate_encryption_decryption(M):
    K = gen_key(symboles)
    decryption_key = gen_decryption_key(K)

    # Encrypt the message
    C = chiffrer(M, K)

    # Decrypt the message
    decrypted_message = decrypt(C, decryption_key)
    return M, C, decrypted_message

# Tester avec un échantillon
M = corpus[10000:10100]
M_original, C, M_decrypted = simulate_encryption_decryption(M)

print(f"Original message M:\n{M_original}")
print(f"Encrypted C:\n{C}")
print(f"Decrypted message M:\n{M_decrypted}")

# Verify correctness
print("Decryption successful:", M == M_decrypted)
````

````python
import random as rnd
import requests
from collections import Counter

# Fonction pour diviser le texte en paires de caractères
def cut_string_into_pairs(text):
    pairs = []
    for i in range(0, len(text) - 1, 2):
        pairs.append(text[i:i + 2])
    if len(text) % 2 != 0:
        pairs.append(text[-1] + '_')  # Ajouter un remplissage si le nombre de caractères est impair
    return pairs

# Fonction pour charger le texte depuis une URL
def load_text_from_web(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors du chargement du texte: {e}")
        return None

# Charger et préparer les symboles
url1 = "https://www.gutenberg.org/ebooks/13846.txt.utf-8"
url2 = "https://www.gutenberg.org/ebooks/4650.txt.utf-8"
corpus = load_text_from_web(url1) + load_text_from_web(url2)

caracteres = list(set(list(corpus)))
nb_caracteres = len(caracteres)
nb_bicaracteres = 256 - nb_caracteres
bicaracteres = [item for item, _ in Counter(cut_string_into_pairs(corpus)).most_common(nb_bicaracteres)]
symboles = caracteres + bicaracteres

# Génération de la clé de substitution
def gen_key(symboles):
    if len(symboles) > 256:
        return False
    rnd.seed(1337)  # Graine pour reproductibilité
    int_keys = rnd.sample(list(range(256)), len(symboles))
    dictionary = {s: "{:08b}".format(k) for s, k in zip(symboles, int_keys)}
    return dictionary

# Création du dictionnaire de chiffrement
K = gen_key(symboles)

# Fonction pour encoder un message en symboles
def M_vers_symboles(M, K):
    encoded_text = []
    i = 0
    while i < len(M):
        if i + 1 < len(M):
            pair = M[i] + M[i + 1]
            if pair in K:
                encoded_text.append(pair)
                i += 2  # Sauter les deux caractères utilisés
                continue
        if M[i] in K:
            encoded_text.append(M[i])
        else:
            encoded_text.append(M[i])
        i += 1
    return encoded_text

# Fonction pour chiffrer un message
def chiffrer(M, K):
    l = M_vers_symboles(M, K)
    l = [K[x] for x in l]
    return ''.join(l)

# Générer un message à chiffrer
M = corpus[10000:10100]  # Segment du texte
C = chiffrer(M, K)

# Fonction pour inverser la clé pour le déchiffrement
def inverser_cle(cle):
    return {v: k for k, v in cle.items()}

# Fonction pour déchiffrer un message chiffré
def dechiffrer(C, cle_inverse):
    symboles_chiffres = [C[i:i+8] for i in range(0, len(C), 8)]
    message_dechiffre = ''.join([cle_inverse.get(symbole, '?') for symbole in symboles_chiffres])
    return message_dechiffre

# Inverser la clé
cle_inverse = inverser_cle(K)

# Déchiffrement du message
message_dechiffre = dechiffrer(C, cle_inverse)

print("Message chiffré :", C)
print("Message déchiffré :", message_dechiffre)

# Verify correctness
print("Decryption successful:", M== message_dechiffre)
````

````python
import unittest
import random
import time
from difflib import SequenceMatcher, unified_diff

def similarity_ratio(str1, str2):
    """
    Calcule le pourcentage de similarité entre deux chaînes de caractères.
    """
    return SequenceMatcher(None, str1, str2).ratio()


def print_diff(original, decrypted):
    """
    Imprime les différences entre le texte original et le texte déchiffré.
    """
    diff = unified_diff(
        original.splitlines(),
        decrypted.splitlines(),
        fromfile='Original',
        tofile='Déchiffré',
        lineterm=''
    )
    for line in diff:
        print(line)

# Fonction pour inverser la clé pour le déchiffrement
def inverser_cle(cle):
    return {v: k for k, v in cle.items()}

# Fonction pour déchiffrer un message chiffré
def dechiffrer(C, cle_inverse):
    symboles_chiffres = [C[i:i+8] for i in range(0, len(C), 8)]
    message_dechiffre = ''.join([cle_inverse.get(symbole, '?') for symbole in symboles_chiffres])
    return message_dechiffre



class TestDecryption(unittest.TestCase):

    def test_decryption_accuracy(self):
        # Charger le premier corpus et enlever les 10 000 premiers caractères
        url1 = "https://www.gutenberg.org/ebooks/13846.txt.utf-8"
        corpus1 = load_text_from_web(url1)
        corpus1 = corpus1[1905:]  # Enlever les 10 000 premiers caractères

        # Charger le deuxième corpus et enlever les 10 000 premiers caractères
        url2 = "https://www.gutenberg.org/ebooks/4650.txt.utf-8"
        corpus2 = load_text_from_web(url2)
        corpus2 = corpus2[3749:]  # Enlever les 10 000 premiers caractères

        # Combiner les deux corpus
        corpus = corpus1 + corpus2

        caracteres = list(set(list(corpus)))
        nb_caracteres = len(caracteres)
        nb_bicaracteres = 256 - nb_caracteres
        bicaracteres = [item for item, _ in Counter(cut_string_into_pairs(corpus)).most_common(nb_bicaracteres)]
        symboles = caracteres + bicaracteres
        nb_symboles = len(symboles)
        dictionnaire = gen_key(symboles)

        a = random.randint(3400, 7200)
        b = random.randint(12000, 15000)
        l = a+b
        c = random.randint(0, len(corpus)-l)

        M = corpus[c:c+l]

        K = gen_key(symboles)
        C = chiffrer(M, K)

        # Inverser la clé
        cle_inverse = inverser_cle(K)


        # Charger le message original M
        original_message = M  # Remplacer par le texte original utilisé pour le chiffrement

        cryptogram = C  # Remplacer par le cryptogramme chiffré

        # Appeler la fonction de déchiffrement de l'étudiant
        decrypted_message = dechiffrer(cryptogram,cle_inverse)

        # Calculer la similarité
        similarity = similarity_ratio(original_message, decrypted_message)
        print(f"Similarité : {similarity:.2%}")

        # Imprimer les différences si la similarité est inférieure à 95 %
        if similarity < 0.985:
            print("Différences entre les messages :")
            print_diff(original_message, decrypted_message)

        # Vérifier que la similarité entre le message original et le message déchiffré est d'au moins 95 %
        self.assertGreaterEqual(similarity, 0.985,
                                f"La similarité est seulement de {similarity:.2%}, ce qui est inférieur à 95 %.")


if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
````
